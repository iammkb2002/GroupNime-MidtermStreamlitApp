# Data Cleaning

Based on the initial exploration, we proceed to clean the data to prepare it for analysis.


## Handling Missing Values

We observed that the 'Arrival Delay in Minutes' column has missing values. Since the missing values are a small fraction of the dataset, we decide to remove these rows.


```python
# Drop rows with missing 'Arrival Delay in Minutes'
df = df.dropna(subset=['Arrival Delay in Minutes'])
```

## Data Type Conversion

We convert categorical columns to the 'category' data type for more efficient memory usage and better analysis.


```python
# Convert categorical columns to 'category' data type
categorical_cols = ['satisfaction', 'Gender', 'Customer Type', 'Type of Travel', 'Class']
for col in categorical_cols:
    df[col] = df[col].astype('category')

# Convert 'Arrival Delay in Minutes' to integer type
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].astype(int)
```

## Check for Duplicates

We check for and remove any duplicate rows to ensure data integrity.


```python
# Remove duplicate rows if any
df = df.drop_duplicates()
```

## Outlier Detection and Handling

We detect and handle outliers in 'Departure Delay in Minutes' and 'Arrival Delay in Minutes' by capping them at the 99th percentile.


```python
# Cap outliers at the 99th percentile
departure_delay_cap = df['Departure Delay in Minutes'].quantile(0.99)
arrival_delay_cap = df['Arrival Delay in Minutes'].quantile(0.99)

df['Departure Delay in Minutes'] = np.where(
    df['Departure Delay in Minutes'] > departure_delay_cap,
    departure_delay_cap,
    df['Departure Delay in Minutes']
)

df['Arrival Delay in Minutes'] = np.where(
    df['Arrival Delay in Minutes'] > arrival_delay_cap,
    arrival_delay_cap,
    df['Arrival Delay in Minutes']
)
```

# Data Exploration

We now perform exploratory data analysis using visualizations to understand the distributions and relationships in the data.

```python
# Histogram of Age
plt.figure(figsize=(10, 4))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Insights from Histogram of Age

- The **Age** distribution is approximately symmetrical with a slight right skew, indicating a slightly larger number of younger passengers.
- The mean age is around **39 years**, with the majority of passengers between **27 and 51 years old**.
- This suggests that the airline caters to a wide range of age groups, but with a concentration in the working-age population.

## Correlation Matrix and Heatmap

We compute the correlation matrix for the numerical variables to observe relationships between them.


```python
# Compute correlation matrix
corr_matrix = df[numerical_cols].corr()

# Display correlation matrix
corr_matrix
```

```python
# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

### Insights from Correlation Matrix and Heatmap

- **Strong Positive Correlation between Departure and Arrival Delays**: The correlation coefficient of approximately **0.95** indicates that flights that depart late tend to arrive late.
- **Negative Correlation between Age and Flight Distance**: A correlation of **-0.25** suggests that younger passengers are more likely to take longer flights, which could be due to travel preferences or business needs.
- **Positive Correlation between Flight Distance and Delays**: Longer flights have a slight tendency to experience more delays, possibly due to increased complexity in operations.
- Understanding these relationships can help the airline in scheduling, resource allocation, and improving customer satisfaction.
